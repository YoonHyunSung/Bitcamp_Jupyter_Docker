```python
def plot_label():
    plt.title("plotting")
    plt.plot([10, 20, 30, 40], label='asc')
    plt.plot([40, 30, 10, 10], label='desc')
    plt.legend()
    plt.show()
```

```python
plot_label()
```

```python
def plot_marker():
    plt.title("plotting")
    plt.plot([10, 20, 30, 40],'r.--', label='circle')
    plt.plot([40, 30, 10, 10],'g^' ,label='triangle up')
    plt.legend()
    plt.show()
```

```python
plot_marker()
```
